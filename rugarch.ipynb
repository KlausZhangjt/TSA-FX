```r
train_data = read.csv("data/train.csv",header=F)
```

```r
library("rugarch");
```

```r
garch_spec=ugarchspec(variance.model=list(garchOrder=c(1,1)),
                 mean.model=list(armaOrder=c(5,5)))
arima_garch=ugarchfit(spec=garch_spec,data=train_data)
arima_garch
```

```r
egarch_spec=ugarchspec(variance.model=list(model="eGARCH",garchOrder=c(1,1)),
                 mean.model=list(armaOrder=c(5,5)))
arima_egarch=ugarchfit(spec=egarch_spec,data=train_data)
arima_egarch
```
